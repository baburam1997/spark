```python
from pyspark.sql import SparkSession
# Create a SparkSession
spark = SparkSession.builder.appName("Working with Different Types of Data").getOrCreate()
```

```python
retail_data = spark.read.format("csv") \
                    .option("header", "true")\
                    .option("inferSchema", "true")\
                    .load("data/data_retail_data_by_day_2011_11_08.csv")
retail_data.show(4070)
```

```python
retail_data.printSchema()
```

# Number

```python
from pyspark.sql.functions import expr, pow,col
fabricatedQuantity = (pow(col("Quantity") * col("UnitPrice"), 2) + 5)
retail_data\
	.select(expr("CustomerId"), fabricatedQuantity.alias("realQuantity"))\
	.show(20)
```

# Boolean

```python
from pyspark.sql.functions import col
retail_data\
    .where(col("InvoiceNo") != 536365)\
    .select("InvoiceNo", "Description")\
    .show(10, False)
# or
retail_data\
    .select("InvoiceNo", "Description")\
    .where("InvoiceNo != 536365")\
    .show(10, False)
```

```python
from pyspark.sql.functions import instr
priceFilter = col("UnitPrice") > 600
descripFilter = instr(retail_data.Description, "POSTAGE") >= 1
retail_data.where(retail_data.StockCode.isin("DOT"))\
	.where(priceFilter | descripFilter).show()
```

```python
from pyspark.sql.functions import instr
DOTCodeFilter = col("StockCode") == "DOT"
priceFilter = col("UnitPrice") > 600
descripFilter = instr(col("Description"), "POSTAGE") >= 1
retail_data\
    .withColumn("isExpensive", DOTCodeFilter & (priceFilter | descripFilter))\
    .where("isExpensive")\
    .select("unitPrice", "isExpensive").show(5)
```

```python
from pyspark.sql.functions import expr
retail_data\
    .withColumn("isExpensive", expr("NOT UnitPrice <= 250"))\
    .where("isExpensive")\
    .select("Description", "UnitPrice").show(10)
```

```python
from pyspark.sql.functions import instr

containsORIGAMI = instr(col("Description"), "ORIGAMI") >= 1
containsINCENSE = instr(col("Description"), "INCENSE") >= 1

retail_data.withColumn("hasSimpleOriInc", containsORIGAMI | containsINCENSE)\
.where("hasSimpleOriInc")\
.select("Description").show(3, False)
```

```python
from pyspark.sql.functions import datediff, months_between, to_date,date_add,date_sub
dateDF.withColumn("week_ago", date_sub(col("today"), 7))\
        .select(datediff(col("week_ago"), col("today"))).show(1)

dateDF.withColumn("week_after", date_add(col("today"), 7))\
        .select(datediff(col("week_after"), col("today"))).show(1)

dateDF.select(to_date(lit("2016-01-01")).alias("start"),
            to_date(lit("2017-05-22")).alias("end"))\
    .select(months_between(col("end"), col("start")))\
    .show(1)
```

# UDF

```python
udfExampleDF = spark.range(5).toDF("num")
def power3(double_value):
    return double_value ** 3
power3(2)
```

```python
udfExampleDF = spark.range(5).toDF("num")
udfExampleDF.show()
```

```python
# we register it:
from pyspark.sql.functions import udf
power3udf = udf(power3)
```

```python
# calling udf
from pyspark.sql.functions import col
udfExampleDF.select(power3udf(col("num"))).show(5)
```

```python
# creating dataframe
udfExample2DF = spark.range(10).toDF("num")

# python funtion
def square(double_value):
    return double_value ** 2

# calling python function
square(2)
```

```python
# register udf
from pyspark.sql.types import IntegerType
spark.udf.register("square_fun", square, IntegerType())
# <function __main__.square(double_value)>

# The function square_fun replaced a previously registered function.
```

```python
# calling udf
udfExample2DF.selectExpr("square_fun(num)").show(10)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, udf,round
from pyspark.sql.types import DoubleType

# Create a SparkSession
spark = SparkSession.builder.getOrCreate()

# Define the UDF to calculate the total price
def calculate_total_price(quantity, unit_price):
    return quantity * unit_price

# Register the UDF and renaming function is optional
# total_price_udf = udf("total_price",calculate_total_price, DoubleType())
total_price_udf = udf(calculate_total_price, DoubleType())

# Load the data with the given schema
retail_data = spark\
    .read\
    .format("csv") \
    .option("header", "true")\
    .option("inferSchema", "true")\
    .load("data/data_retail_data_by_day_2011_11_08.csv")

# Apply the UDF to create a new column
data_with_total_price = retail_data\
    .withColumn("TotalPrice", total_price_udf(col("Quantity"), col("UnitPrice")))

# Show the result
data_with_total_price\
    .select("Quantity","UnitPrice",round("TotalPrice",2).alias("TotalPrice"))\
    .show()
```
